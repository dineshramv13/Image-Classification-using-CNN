```python
y_train_categorical = keras.utils.to_categorical(
    y_train, num_classes=10 , dtype = 'float32'
)
```

```python
y_test_categorical = keras.utils.to_categorical(
    y_test, num_classes=10, dtype='float32'
)
```

```python
y_train[0:5]
```

```python
y_train_categorical[0:5]
```

```python
model = keras.Sequential([
        keras.layers.Flatten(input_shape=(32,32,3)),
        keras.layers.Dense(3000, activation='relu'),
        keras.layers.Dense(1000, activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')    
    ])

model.compile(optimizer='SGD',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

```

```python
model.fit(x_train_scaled, y_train_categorical, epochs=5)
```

```python
np.argmax(model.predict(x_test_scaled)[1])
```

```python
classes[y_test[1][0]]
```

```python
model.evaluate(x_test_scaled , y_test_categorical)
```

```python

from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = model.predict(x_test_scaled)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
classes =['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
def preprocess_custom_image(img_path):
    img = Image.open(img_path)  # Read the image
    img = img.resize((32, 32))  # Resize the image to 28x28
    #img = img.convert('L')  # Convert the image to grayscale
    img = np.array(img)  # Convert to numpy array
    img = img.astype('float32') / 255.0  # Normalize pixel values
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img
# Path to your custom image
custom_img_path = 'truck.jpeg'

# Preprocess the custom image
custom_img = preprocess_custom_image(custom_img_path)
```

```python
# Predict the class of the custom image
predictions = Testing_saved_model.predict(custom_img)
predicted_class = np.argmax(predictions)

# Display the prediction result
plt.imshow(custom_img.squeeze(), cmap='gray')
plt.title(f"Predicted : {classes[predicted_class]}")  # No need to subtract or add anything
plt.axis('off')
plt.show()
```

## CNN

```python
cnn = keras.models.Sequential([
    keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    keras.layers.MaxPooling2D((2, 2)),
    
    keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
cnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(x_train_scaled, y_train_categorical, epochs=5)
```

```python
np.argmax(cnn.predict(x_test_scaled)[1])
```

```python
classes[y_test[1][0]]
```

```python
cnn.evaluate(x_test_scaled , y_test_categorical)
```

```python

from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = cnn.predict(x_test_scaled)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```
